```python
directory = '/Users/bruno/Library/Mobile Documents/com~apple~CloudDocs/Thesis/Jupter Notebooks'

df_eps_re = pd.read_csv(directory + '/190418/epseffreal.csv')
df_eps_im = pd.read_csv(directory + '/190418/epseffimag.csv')

log10tandelta = np.array(df_eps_re['aSiloss De-Embedded Frequency=5.0GHz '].index)[1::].astype('float')
eps_r = np.array(df_eps_re['aSiloss De-Embedded Frequency=5.0GHz '])[1::].astype('float')
eps_im = np.array(df_eps_im['aSiloss De-Embedded Frequency=5.0GHz '])[1::].astype('float')
```

```python
eps = eps_r + 1j*eps_im
freq = 5*10**9

tandelta = -(27.3/np.pi)*np.tan(np.angle(eps))
tandelta[np.logical_not(tandelta > 0)] = np.nan

tandelta_di = 10**log10tandelta
```

```python
fig = plt.figure()
ax = plt.gca()

ax.scatter(log10tandelta,np.log10(tandelta))
ax.set_xlim([-8,-1])
ax.set_xlabel('(10log) Dielectric loss tangent')
ax.set_ylabel('(10log) Loss tangent of CPW')
ax.set_title('Loss tangent of Al CPW on a-Si, calculated in Sonnet')

fig2 = plt.figure()
ax2 = plt.gca()

ax2.scatter(log10tandelta,np.log10(1/tandelta), c='orange')
ax2.set_xlim([-8,-1])
ax2.set_xlabel('(10log) Dielectric loss tangent')
ax2.set_ylabel('(10log) Q_i of CPW')
ax2.set_title('Q_i of Al CPW on a-Si, calculated in Sonnet')
```

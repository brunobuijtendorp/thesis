## Results of May 20
Laser power 1%, 100s exposure, 1 accumulation, 100x objective.

```python
data0520_100C = np.flipud(np.loadtxt(data_dir + '20190520_100C_300s_1%_1acc_obj100.txt'))
data0520_250C = np.flipud(np.loadtxt(data_dir + '20190520_250C_300s_1%_1acc_obj100.txt'))
data0520_350C = np.flipud(np.loadtxt(data_dir + '20190520_350C_300s_1%_1acc_obj100.txt'))

popt100C, ax = fit(data0520_100C[:,0], data0520_100C[:,1])
ax.set_title("$T_{sub}$ 100 °C")

popt250C, ax = fit(data0520_250C[:,0], data0520_250C[:,1])
ax.set_title("$T_{sub}$ 250 °C")

popt350C, ax = fit(data0520_350C[:,0], data0520_350C[:,1])
ax.set_title("$T_{sub}$ 350 °C")
```

```python
I_100C, dtheta_I_100C = relative_intensity(popt100C)
I_250C, dtheta_I_250C = relative_intensity(popt250C)
I_350C, dtheta_I_350C = relative_intensity(popt350C)
dtheta_omega_100C = (popt100C[9] - 505.5)/(-2.5)
dtheta_omega_250C = (popt250C[9] - 505.5)/(-2.5)
dtheta_omega_350C = (popt350C[9] - 505.5)/(-2.5)
```

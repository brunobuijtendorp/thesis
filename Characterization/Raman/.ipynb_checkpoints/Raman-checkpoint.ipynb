# Raman measurements

```python
import numpy as np
import matplotlib.pyplot as plt
import os
data_dir = os.getcwd() + '/data/'
```

## Results of May 20
Laser power 1%, 100s exposure, 1 accumulation, 100x objective.

```python
data0520_100C = np.flipud(np.loadtxt(data_dir + '20190520_100C_300s_1%_1acc_obj100.txt'))
data0520_250C = np.flipud(np.loadtxt(data_dir + '20190520_250C_300s_1%_1acc_obj100.txt'))
data0520_350C = np.flipud(np.loadtxt(data_dir + '20190520_350C_300s_1%_1acc_obj100.txt'))

fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize=(16.5,18))
ax1.plot(data0520_100C[:,0], data0520_100C[:,1])
ax1.set_title("$T_{sub}$ 100 °C")
ax1.set_xlabel(r'Wavenumber ($cm^{-1}$)')
ax1.set_ylabel('Intensity (a.u.)')
ax1.axvline(150, linestyle='--', color = 'k', alpha = 0.2)
ax1.axvline(480, linestyle='--', color = 'k', alpha = 0.2)
ax1.axvline(520, linestyle='--', color = 'k', alpha = 0.2)
ax1.autoscale(enable=True) 

ax2.plot(data0520_250C[:,0], data0520_250C[:,1])
ax2.set_title("$T_{sub}$ 250 °C")
ax2.set_xlabel(r'Wavenumber ($cm^{-1}$)')
ax2.set_ylabel('Intensity (a.u.)')
ax2.axvline(150, linestyle='--', color = 'k', alpha = 0.2)
ax2.axvline(480, linestyle='--', color = 'k', alpha = 0.2)
ax2.axvline(520, linestyle='--', color = 'k', alpha = 0.2)

ax3.plot(data0520_350C[:,0], data0520_350C[:,1])
ax3.set_title("$T_{sub}$ 350 °C")
ax3.set_xlabel(r'Wavenumber ($cm^{-1}$)')
ax3.set_ylabel('Intensity (a.u.)')
ax3.axvline(150, linestyle='--', color = 'k', alpha = 0.2)
ax3.axvline(480, linestyle='--', color = 'k', alpha = 0.2)
ax3.axvline(520, linestyle='--', color = 'k', alpha = 0.2)
```

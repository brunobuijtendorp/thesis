# Raman measurements
Useful resource: Determining the material structure of microcrystalline silicon from Raman spectra, Smit et al. (2003)
## Silicon peaks

### a-Si peaks
* 150 $cm^{-1}$: transverse acoustic (TA) a-Si peak
* 480 $cm^{-1}$: transverse optic (TO) a-Si peak

### Other peaks

* 520 $cm^{-1}$: crystalline silicon peak
* 650 $cm^{-1}$: two-phonon excitations in a-Si:H
* 950 $cm^{-1}$: Si-H wagging mode
* 960 $cm^{-1}$: two-phonon excitations in c-Si


## What we can learn from the spectrum
### Bond angle disorder $\Delta\theta$ 
Source: https://dspace.library.uu.nl/bitstream/handle/1874/680/c6.pdf

There is a relation between $\Delta\theta$ and:
* The TO peak width $\Gamma$
* The TO peak position $\omega_{TO}$
* The TO/TA intensity ratio I

### Amorphicity
Source: Investigation of E-Beam Evaporated Silicon Film Properties for MEMS Applications, Michael et al. (2015)

If we define $A_1$, $A_2$, and $A_3$ to represent the areas of the Lorentzian profile at 520 $cm^{-1}$, and the Gaussian profiles at 500 $cm^{-1}$ and 480 $cm^{-1}$, respectively, then the crytalline fraction $C$ is:

$C = \frac{A_1 + A_2}{A_1 + A_2 + 0.8A_3}.$

### Quick observations (to get a feeling)
* A wider $\Gamma$ peak corresponds to a larger $\Delta\theta$
* The amorphous peak shifts to higher frequencies with higher crystallinity
